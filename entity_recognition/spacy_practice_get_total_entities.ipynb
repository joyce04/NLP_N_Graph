```python
import pandas as pd
```

```python
df = pd.read_csv("../topic_modeling/titles_condition.tsv", sep='\t')
df['length'] = df['Analysis of efficacy'].apply(lambda x: len(str(x)))
df.head()
```

```python
df.shape
```

```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load('en')
```

```python
max_limit = 900000
cum_len = 0
start_ind = 0
ent_list = []
    
for ind, row in df.iterrows():
    cum_len += row.length
    if cum_len >= max_limit:
        sub_text = df.iloc[start_ind:ind]['Analysis of efficacy']
        start_ind = ind
        cum_len = 0
        doc = nlp(' '.join(sub_text.tolist()))

        for ent in doc.ents:
            if len(ent.text)>2:
                ent_list.append({'cluster':1,\
                                'text':ent.text, \
                                'label':ent.label_})
        print('*********************************')
```

```python
pd.DataFrame(ent_list).groupby(['cluster', 'label']).count().reset_index().to_dict(orient='records')
```

```python
ent_df = pd.DataFrame(ent_list)
ent_df.head()
```

```python
ent_df.groupby(ent_df.label).count()
```
